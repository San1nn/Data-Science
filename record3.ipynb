```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,8))


plt.subplot(3,3,1)
#1st graph
x=np.array([2,8])
y=np.array([20,30])

plt.scatter(x,y)
plt.xlabel("x label",c="blue")
plt.ylabel("y label",c="yellow")
plt.title("TITLE 1")


plt.subplot(3,3,2)
#2nd graph
x=np.array([2,3,4,5,6])
y=np.array([22,28,22,28,22])
plt.plot(x,y,marker="*",ms="15",color="orange",linewidth="2.5")

x=np.array([2,3,4,5,6])
y=np.array([26,23,26,23,26])
plt.plot(x,y,linestyle="dashed",color="blue")


plt.subplot(3,3,3)
#3rd graph
x=np.array([2,3,4,5])
y=np.array([23,26,24,27])
plt.plot(x,y,c="blue",marker="o",ms="15",mfc="yellow",mec="black")


plt.subplot(3,3,4)
#4th graph
plt.title("TITLE 2")
x=np.array([2,8])
y=np.array([20,30])
plt.grid()
plt.plot(x,y)
plt.tight_layout()


plt.subplot(3,3,5)
#5th graphaddition
x=np.array([2,3,4,5,6])
y=np.array([22,23,24,22,23])
colors=np.array(["blue","green","yellow","red","orange"])
plt.scatter(x,y,c=colors,alpha=0.5)

#6th
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.subplot(3,3,6)
plt.bar(x,y,width=0.1,color='r')
plt.show()


#7th
y=np.array([35,20,20,25])
label=["BIKE","CAR","BUS","SHIP"]
myexplode=[0.2,0,0,0]
color=["black",'r','y','c']

#8th
plt.pie(y,startangle=90,explode=myexplode,colors=color,labels=label,shadow=True)
plt.legend()
plt.show()


plt.show()
```
